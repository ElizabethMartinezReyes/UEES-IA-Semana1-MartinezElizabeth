# Introducción a Deep Learning con TensorFlow/Keras

Crearemos una red neuronal simple y visualizaremos su entrenamiento.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Datos de ejemplo
X = np.array([[0], [1], [2], [3], [4], [5]], dtype=float)
y = np.array([[0], [1], [2], [3], [4], [5]], dtype=float)

# Crear modelo
model = Sequential([
    Dense(10, input_shape=(1,), activation='relu'),
    Dense(1)
])

# Compilar y entrenar
model.compile(optimizer='adam', loss='mse')
hist = model.fit(X, y, epochs=200, verbose=0)
print("Entrenamiento finalizado")
```

## Visualización del entrenamiento

```python
import matplotlib.pyplot as plt
plt.plot(hist.history['loss'])
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.show()
```
