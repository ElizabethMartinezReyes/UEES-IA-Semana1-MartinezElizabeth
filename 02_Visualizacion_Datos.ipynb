# Visualización de Datos con Matplotlib, Seaborn y Plotly

Este notebook explora distintas librerías de visualización en Python.

## Gráficos con Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Gráfico de Seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

## Visualización estadística con Seaborn

```python
import seaborn as sns
import pandas as pd

# Dataset de ejemplo
df = sns.load_dataset('iris')
sns.pairplot(df, hue='species')
```

## Gráficos interactivos con Plotly (Bonus)

```python
import plotly.express as px

# Usar el mismo dataset de iris
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species', title='Gráfico interactivo')
fig.show()
```
