# Fundamentos de NumPy y Pandas

Este notebook contiene ejemplos básicos de manipulación de arrays con NumPy y análisis de datos con Pandas usando el dataset **Titanic** de Seaborn.

## Manipulación de arrays con NumPy

```python
import numpy as np

# Crear un array 1D
a = np.array([1, 2, 3, 4, 5])
print("Array 1D:", a)

# Operaciones
print("Suma:", a + 2)
print("Multiplicación:", a * 3)
print("Media:", np.mean(a))
print("Forma del array:", a.shape)
```

## Análisis de datos con Pandas (usando dataset Titanic)

```python
import pandas as pd
import seaborn as sns

# Cargar dataset Titanic
df = sns.load_dataset('titanic')
df.head()
```

### Información general del dataset

```python
df.info()
```

### Estadísticas descriptivas

```python
df.describe(include='all')
```

### Filtrar pasajeros que sobrevivieron y eran mujeres

```python
mujeres_sobrevivientes = df[(df['sex'] == 'female') & (df['survived'] == 1)]
mujeres_sobrevivientes[['sex', 'age', 'class', 'survived']].head()
```

### Agrupar por clase y calcular tasa de supervivencia

```python
df.groupby('class')['survived'].mean()
```
